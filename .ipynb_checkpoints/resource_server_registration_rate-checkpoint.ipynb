```python
serverResourceDf = resourceDf[
    resourceDf["type"].isin(['server','legacy-server','physical-legacy-server'])]
serverResourceDf.head(2)
```

```python
import json
defaultPA = json.loads(paDf[paDf["name"]=='default'].to_json())
print(defaultPA['_id']['0']['$oid'])
```
